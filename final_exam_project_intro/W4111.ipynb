__Question:__ Create a view ```employee_customer_sales``` with the following information:
- ```employeeNumber```
- ```employeeLastname```
- ```employeeFirstName```
- ```customerNumber```
- ```customerName```
- ```revenue```

- The employee information is for the employee that is the ```customer.customerRepEmployeeNumber```.
- ```revenue``` is the total revenue over all of the customer's orders.
    - The revenue for an ```order``` is ```priceEach*quantityOrdered``` for each ```orderdetails``` in the order.

__Answer:__


````python
%%sql
use classicmodels;
drop view if exists employee_customer_sales;
create view employee_customer_sales as(
with table1 as(
    select orders.customerNumber,
           sum(orderdetails.quantityOrdered * orderdetails.priceEach) 
               as revenue
    from orderdetails 
        join orders 
            using(orderNumber) 
    group by customerNumber
), table2 as (
    select salesRepEmployeeNumber as employeeNumber,
           customerNumber,
           customers.customerName,
           revenue
    from table1 
        join customers 
            using(customerNumber)
), table3 as (
    select employees.employeeNumber,
           lastName  as employeeLastname,
           firstName as employeeFirstname,
           customerNumber,
           customerName,
           revenue
    from table2 
        inner join employees 
            on table2.employeeNumber = employees.employeeNumber
)
select * 
from table3 
order by employeeLastname asc);
````

## S3

__Question:__ 


- Use the ```customers``` and ```orders``` for this query.


- Shipping days is the number of days between ```orderDate``` and ```shippedDate.```


- Product a table of the form:
    - ```customerNumber```
    - ```customerName```
    - ```noOfOrders``` is the number of orders the customer placed.
    - ```averageShippingDays```, which is the average shipping days.
    - ```minimumShippingDays```, which is the minimum shipping days.
    - ```maximumShippingDays```, which is the maximum shipping days.
    
    
- The table should only contain entries where:
    - ```noOfOrders >= 3```
    - ```averageShippingDays >= 5``` or ```maximumShippingDays >= 10.```
    

__Answer:__

````python
%%sql

with table1 as (
    select customerNumber,
           avg(DATEDIFF(shippedDate, orderDate)) as averageShippingDays,
           min(DATEDIFF(shippedDate, orderDate)) as minimumShippingDays,
           max(DATEDIFF(shippedDate, orderDate)) as maximumShippingDays
    from orders 
    group by customerNumber
), table2 as(
    select customerNumber,
           count(*) as noOfOrders
    from orders
    group by customerNumber
), table3 as (
    select customerNumber,
           noOfOrders,
           averageShippingDays,
           minimumShippingDays,
           maximumShippingDays
    from table1
        join table2
            using (customerNumber)
    where (noOfOrders >= 3 and averageShippingDays >= 5)
       or (noOfOrders >= 3 and maximumShippingDays >= 10)
)
select customerNumber,
       customerName,
       noOfOrders,
       averageShippingDays,
       minimumShippingDays,
       maximumShippingDays
from table3
    join customers
        using(customerNumber);
````

## Q3

__Question:__

Briefly explain:
- Logical block addressing, CHS addressing
- RAID-0, RAID-1, RAID-5
- Fixed length records, variable length records.

__Answer:__

Logical block addressing (LBA) is a way of specifying the location of data on a storage device, such as a hard drive or solid-state drive. In LBA, each block of data on the storage device is assigned a unique numerical address, which can be used to quickly locate and access the desired data. LBA is a more efficient and flexible addressing scheme compared to other methods, such as cylinder-head-sector (CHS) addressing. CHS addressing is a method of "specifying the location of data on a storage device that uses the geometry of the device to identify the location of a particular block of data." In CHS addressing, the location of a block of data is specified using the cylinder, head, and sector numbers of the disk platter where the data is stored. This method of addressing is less efficient and less flexible than LBA, and is therefore not commonly used in modern computing systems. RAID 0, also known as striping, is a technique for increasing the performance of a storage system by spreading data across multiple disks. RAID 1, also known as mirroring, is a technique for providing fault tolerance by storing multiple copies of data on multiple disks. RAID 5, also known as striping with parity, is a technique for providing fault tolerance and improved performance by storing data across multiple disks, with parity information distributed among the disks. Fixed-length records are records that have a fixed number of fields, and each field has a fixed length. Variable-length records are records that have a variable number of fields, and the length of each field can vary.

__Question:__

Briefly explain:
- Equivalent queries
- Hash Join
- Materialization, Pipelining

__Answer:__

Equivalent queries are SQL queries that return the same result set, but may have different syntax or structure. For example, a query that returns the sum of a column in a table can be written in many different ways, but they all return the same result. A hash join is a type of join operation in relational databases that uses a hash table to build a join. It involves building a hash table "for each of the two relations being joined, and then looking up matching tuples in the hash table." Hash joins are most efficient when one relation is much smaller than the other, as the hash table for the smaller relation can be held in memory. If one relation is very small and the other is very large, this can be a very efficient and fast way to join the two relations. Materialization and pipelining are two techniques used to optimize the execution of a query. Materialization involves "storing the results of a query in a temporary table, which can then be used to answer subsequent queries." Pipelining involves "breaking down a query into intermediate steps," and executing each step in turn. This can reduce the amount of data that needs to be read, and also "allow intermediate results to be shared between multiple queries."

--Reference from textbook, https://www.oreilly.com/library/view/high-performance-mysql/9780596101718/ch04.html
